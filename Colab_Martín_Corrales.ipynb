````python
# Datos proporcionados
categorias = ['sin_victima', 'con_victima']
valores = [108802, 55046]

plt.figure(figsize=(8, 6))
plt.bar(categorias, valores, color=['blue', 'red'])
plt.xlabel('Categorías')
plt.ylabel('Cantidad de ocurrencias')
plt.title('Distribución de ocurrencias: sin_victima vs con_victima')
plt.show()
````

Se presenta un gráfico de barras que ilustra la distribución de ocurrencias en dos categorías: 'sin_victima' y 'con_victima'. Los datos muestran que hay 108,802 ocurrencias clasificadas como 'sin_victima' y 55,046 ocurrencias clasificadas como 'con_victima'. El gráfico destaca la diferencia en la frecuencia de estas dos categorías.

````python
# Datos proporcionados
total_ocurrencias = 108802 + 55046  # Suma de ambas ocurrencias

# Calculando los porcentajes
porcentaje_sin_victima = (108802 / total_ocurrencias) * 100
porcentaje_con_victima = (55046 / total_ocurrencias) * 100

categorias = ['sin_victima', 'con_victima']
valores_porcentuales = [porcentaje_sin_victima, porcentaje_con_victima]

plt.figure(figsize=(8, 6))
bars = plt.bar(categorias, valores_porcentuales, color=['blue', 'red'])

# Agregar etiquetas con los porcentajes en cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}%', ha='center', va='bottom')

plt.xlabel('Categorías')
plt.ylabel('Porcentaje de ocurrencias')
plt.title('Porcentaje de ocurrencias: sin_victima vs con_victima')
plt.ylim(0, 100)  # Establecer límites del eje y de 0 a 100
plt.show()
````

#Matriz de Correlación

````python

# Calcular la matriz de correlación
correlation_matrix = datos_filtrados.corr()

# Crear un mapa de calor (heatmap) con seaborn
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación')
plt.show()
````

Matriz de Correlación:

La matriz de correlación muestra cómo las variables numéricas en un DataFrame están relacionadas entre sí.
Los valores de la matriz varían entre -1 y 1, donde -1 indica una correlación negativa perfecta, 1 indica una correlación positiva perfecta y 0 indica ausencia de correlación.
Cálculo de la Matriz de Correlación:

Puedes calcular la matriz de correlación en pandas utilizando el método corr() en tu DataFrame.
Visualización con Mapa de Calor:

Se utiliza seaborn y matplotlib para visualizar la matriz de correlación en forma de mapa de calor.
El mapa de calor asigna colores más intensos a valores de correlación más altos, lo que facilita la identificación de patrones.
Cuidados y Consideraciones:

La correlación no implica causalidad; es importante interpretar los resultados con precaución.
La visualización es más efectiva con un número limitado de variables para evitar la congestión visual.
Puedes explorar relaciones específicas utilizando gráficos de dispersión o gráficos de regresión.

````python
# Calcular el total de eventos por mes
total_eventos_por_mes = datos_filtrados.groupby('meses').size()

# Calcular el porcentaje de ilesos y muertos por mes
porcentaje_ilesos_por_mes = (datos_filtrados[datos_filtrados['ilesos'] > 0].groupby('meses').size() / total_eventos_por_mes) * 100
porcentaje_muertos_por_mes = (datos_filtrados[datos_filtrados['mortos'] > 0].groupby('meses').size() / total_eventos_por_mes) * 100

# Crear un histograma de los porcentajes de ilesos y muertos por mes
plt.figure(figsize=(10, 6))
plt.bar(porcentaje_ilesos_por_mes.index, porcentaje_ilesos_por_mes, color='skyblue', label='Ilesos', alpha=0.7)
plt.bar(porcentaje_muertos_por_mes.index, porcentaje_muertos_por_mes, color='red', bottom=porcentaje_ilesos_por_mes, label='Muertos', alpha=0.7)

# Mostrar los porcentajes en las barras
for x, y1, y2 in zip(porcentaje_ilesos_por_mes.index, porcentaje_ilesos_por_mes, porcentaje_muertos_por_mes):
    plt.text(x, y1/2, f"{y1:.2f}%", ha='center', va='center', color='black')
    plt.text(x, y1 + y2/2, f"{y2:.2f}%", ha='center', va='center', color='black')

plt.title('Porcentaje de Ilesos y Muertos por Mes')
plt.xlabel('Mes')
plt.ylabel('Porcentaje')
plt.xticks(range(1, 13))  # Establecer los ticks del eje x como los meses del año
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

````

#Curva de Aprendizaje

````python

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Genera una curva de aprendizaje para un estimador dado.

    Parámetros:
    estimator : objeto estimador
        El modelo a evaluar.
    title : str
        Título del gráfico.
    X : matriz de características
        Datos de entrada.
    y : vector de destino
        Etiquetas de clase correspondientes a los datos de entrada.
    ylim : tuple, opcional
        Límites de la gráfica.
    cv : int, validación cruzada, generador o iterador, opcional
        Estrategia de validación cruzada.
    n_jobs : int o None, opcional (default=None)
        Número de trabajos a ejecutar en paralelo.
    train_sizes : array-like, opcional
        Secuencia de tamaños de entrenamiento.
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Tamaño del conjunto de entrenamiento")
    plt.ylabel("Puntaje")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Puntaje de entrenamiento")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Puntaje de validación cruzada")

    plt.legend(loc="best")
    return plt

# Usar la función para generar la curva de aprendizaje
title = "Curva de aprendizaje (Regresión Logística)"
estimator = LogisticRegression()  # Suponiendo que ya has definido y ajustado tu modelo
plot_learning_curve(estimator, title, X, y, cv=5, n_jobs=-1)

plt.show()
````

Curva de Aprendizaje: Esta gráfica muestra cómo el rendimiento de un modelo varía en función del tamaño del conjunto de entrenamiento. El eje x representa el tamaño del conjunto de entrenamiento, mientras que el eje y representa el puntaje del modelo, que puede ser la precisión, el error, u otra métrica de rendimiento. La curva de aprendizaje está compuesta por dos líneas: una para el puntaje de entrenamiento y otra para el puntaje de validación cruzada.

Interpretación de la curva de aprendizaje: A medida que el tamaño del conjunto de entrenamiento aumenta, el puntaje de entrenamiento y el puntaje de validación cruzada pueden variar. El puntaje de entrenamiento tiende a disminuir a medida que el conjunto de entrenamiento se hace más grande, ya que el modelo encuentra más datos para ajustarse. Por otro lado, el puntaje de validación cruzada tiende a aumentar a medida que el conjunto de entrenamiento se hace más grande, ya que el modelo generaliza mejor con más datos. La distancia entre las dos líneas indica la cantidad de sobreajuste o subajuste del modelo.

Uso de la curva de aprendizaje: Esta visualización ayuda a identificar si el modelo está sobreajustado o subajustado. Un espacio grande entre las dos líneas indica que el modelo está sobreajustado y necesita más datos de entrenamiento para generalizar mejor. Por otro lado, si las dos líneas están cerca una de la otra y muestran un alto puntaje, indica que el modelo está bien ajustado.

En resumen, la curva de aprendizaje proporciona información valiosa sobre cómo el rendimiento del modelo varía con el tamaño del conjunto de entrenamiento y ayuda a tomar decisiones informadas sobre la mejora del modelo.

#Gráfico de dispersión

````python
# Graficar los componentes principales en un gráfico de dispersión bidimensional
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Gráfico de Dispersión de Componentes Principales')
plt.grid(True)
plt.show()
````

Graficar los componentes principales en un gráfico de dispersión bidimensional puede ayudarte a visualizar la distribución de los datos en el nuevo espacio de características y a identificar posibles agrupamientos o patrones.

#Gráfico de carga de componentes principales


````python
# Obtener las cargas de las variables originales en los componentes principales
loadings = pca.components_

# Crear un gráfico de barras para mostrar las cargas de las variables en el primer componente principal
plt.figure(figsize=(10, 6))
plt.bar(range(len(loadings[0])), loadings[0])
plt.xticks(range(len(loadings[0])), X.columns, rotation=90)
plt.xlabel('Variables Originales')
plt.ylabel('Carga en el Primer Componente Principal')
plt.title('Gráfico de Carga del Primer Componente Principal')
plt.show()
````

Este gráfico muestra la contribución de cada variable original a los componentes principales. Puedes trazar las cargas de los componentes principales para identificar qué variables tienen mayor influencia en la estructura de los datos.

#Varianza explicada acumulada



````python
# Calcular la varianza explicada acumulada
explained_variance_ratio_cumulative = np.cumsum(pca.explained_variance_ratio_)

# Crear un gráfico de línea para mostrar la varianza explicada acumulada
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio_cumulative) + 1), explained_variance_ratio_cumulative, marker='o', linestyle='-')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Varianza Explicada Acumulada por Componente Principal')
plt.grid(True)
plt.show()
````

Graficar la varianza explicada acumulada por cada componente principal te ayuda a entender cuánta información se conserva al considerar un número creciente de componentes principales. Esto puede ayudarte a decidir cuántos componentes principales seleccionar para tu análisis.

#Biplot

````python
# Realizar PCA en los datos
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Calcular las cargas de las variables en los componentes principales
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Graficar las observaciones y las cargas de las variables en el biplot
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
for i, variable in enumerate(X.columns):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='r', alpha=0.5)
    plt.text(loadings[i, 0]*1.1, loadings[i, 1]*1.1, variable, color='r', ha='center', va='center')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Biplot')
plt.grid(True)
plt.show()
````

Un biplot combina un gráfico de dispersión de componentes principales con las cargas de las variables originales en el mismo espacio. Esto te permite ver tanto la distribución de las observaciones como la relación entre las variables originales y los componentes principales.
